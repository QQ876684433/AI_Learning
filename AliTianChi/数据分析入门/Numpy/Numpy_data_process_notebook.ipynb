```python
dt = np.dtype([('name', 'S10'), ('age', int)])
a = np.array([("raju", 21), ("anil", 25), ("ravi", 17), ("amar", 27)], dtype=dt)
print('')
print(a)
print()
print('name')
print(a.sort(order='name'))
```

```python
x = np.array([3,  1,  2])  
print ('')  
print (x) 
print ('\n')  
print (' x  argsort() ' )
y = np.argsort(x)  
print (y) 
print ('\n')  
print ('')  
print (x[y])  
print ('\n')  
print ('')  
for i in y:  
    print (x[i])
```
