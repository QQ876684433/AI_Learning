```python
import random
import matplotlib.pyplot as plt

%matplotlib inline

position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)

plt.plot(walk[:])
```

```python
plt.plot(walk[:200])
```

```python
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
```

```python
print(walk)
print(10*'=')
print(walk.min())
print(10*'=')
print(walk.max())
```

```python
plt.plot(walk[:])
```

```python
print((np.abs(walk)>=10).argmax())
```

```python
print((np.abs(walk)>=10))
```

```python
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1)

print(walks)
```

```python
print(walks.max())
print(walks.min())
```

```python
hits30 = (np.abs(walks) > 30).any(1)
print(hits30)
print(hits30.sum())
```

```python
hits40=(np.abs(walks)>=40).any(1)
print(hits40)
print(hits40.sum())
```

```python
hits50=(np.abs(walks)>=50).any(1)
print(hits50)
print(hits50.sum())
```

```python
hits10=(np.abs(walks)>=10).any(1)
print(hits10)
print(hits10.sum())
```

```python
crossing_times = (np.abs(walks) >= 30).argmax()
print(crossing_times)
print(crossing_times.mean())
```
