```python
from matplotlib import pyplot as plt

%matplotlib inline

plt.bar([0.25, 1.25, 2.25, 3.25, 4.25], [50, 40, 70, 80, 80], label='BMW', color='b', width=.25)
plt.bar([0.50, 1.50, 2.50, 3.50, 4.50], [60, 50, 30, 70, 90], label='DaZhong', color='g', width=.25)
plt.bar([0.75, 1.75, 2.75, 3.75, 4.75], [80, 20, 20, 50, 60], label='Audi', color='r', width=.25)

plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Information')
plt.show()
```
