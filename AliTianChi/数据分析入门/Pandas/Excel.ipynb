```python
import pandas as pd, numpy as np

dic2 = {'Name': ['Steve', 'Amy', 'Saar', 'CCL', 'Chph', 'Sue', 'David', 'Jimmy'],
        'Score': [88, 76, 89, 67, 79, 90, 92, 100]}
score = pd.DataFrame(dic2)
print(score)

print(10 * '=')

student = pd.DataFrame({'Age': [11, 23, 51, 2, 1, 45, 63, 21], 'Height': [1, 2, 3, 6, 5, 4, 7, 8],
                        'Weight': [25, 14, 36, 85, 47, 98, 74, 51],
                        'Sex': ['F', 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
                        'Name': ['Steve', 'Jobs', 'Mark', 'CCL', 'Chph', 'Sue', 'David', 'Jimmy']})
print(student)

table1 = pd.pivot_table(student, values=['Height'], columns=['Sex'])
print(table1)
```

```python
table1 = pd.pivot_table(student, values=['Weight'], columns=['Sex'], aggfunc='sum')
print(table1)
```

```python
table2 = pd.pivot_table(student, values=['Weight', 'Height'], columns=['Sex'])
print(table2)
```

```python
table3 = pd.pivot_table(student, values=['Weight', 'Height'], columns=['Sex', 'Age'])
print(table3)
```

```python
table3.unstack()
```

```python
table4 = pd.pivot_table(student, values=['Height', 'Weight'], columns=['Sex', 'Age'],
                        aggfunc=[np.mean, np.median, np.std])
table4.unstack()
table4
```
