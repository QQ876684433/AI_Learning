```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)


t1 = np.arange(0.0, 5.0, 1.0)
t2 = np.arange(0.0, 5.0, 0.02)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2))
plt.subplot(222)
plt.plot(t2, np.cos(2 * np.pi * t2))
plt.subplot(224)
plt.plot(t1, np.exp(t1))
plt.subplot(223)
plt.plot(t2[1:], np.log2(t2[1:]))
plt.show()
```
